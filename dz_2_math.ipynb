```python
def func1(x):
    y = (3/4)*x - 1
    return y

def func2(x):
    y = (3/4)*x + 2
    return y

def func3(x):
    y = 2-2*x
    return y

def func4(x):
    y = (1/4)*(x+1)*(x+1)-2
    return y

def func5(x):
    y = (1/4)*(x-1)*(x-1)-2
    return y

def func6(x):
    y = (x-1)*(x-3)
    return y

def func7(x):
    y = (x+1)*(x+3)
    return y

def func8(x):
    y = 2-x*x
    return y

def func9(x):
    y = 2-(1/4)*x*x
    return y
```

```python
def figure_made(dec):
    x1 = np.linspace(-3.0, 3.0, 100)
    # на основе вектора x строим значения для вектора y
    if dec == 1:
        y1 = [func1(x1[i]) for i in range(len(x1))]
    elif dec == 2:
        y1 = [func2(x1[i]) for i in range(len(x1))]
    elif dec == 3:
        y1 = [func3(x1[i]) for i in range(len(x1))]
    elif dec == 4:
        y1 = [func4(x1[i]) for i in range(len(x1))]
    elif dec == 5:
        y1 = [func5(x1[i]) for i in range(len(x1))]
    elif dec == 6:
        y1 = [func6(x1[i]) for i in range(len(x1))]
    elif dec == 7:
        y1 = [func7(x1[i]) for i in range(len(x1))]
    elif dec == 8:
        y1 = [func8(x1[i]) for i in range(len(x1))]
    elif dec == 9:
        y1 = [func9(x1[i]) for i in range(len(x1))]
        

    plt.plot(x1, y1)
    plt.title('График функции')
    plt.axvline(0, color='r')
    plt.axhline(0, color='r')
    plt.ylabel('y')
    plt.xlabel('x')

    plt.show()
```

```python
x=1
while x != 9:
    figure_made(x)
    x+=1
```

**Задание 4.** Найти минимальную степень следующего полинома:

![alt text](https://pp.userapi.com/c855532/v855532908/aadcd/PZdPYTQT02A.jpg)

```python
x = np.linspace(-2.0, 2.0, 100)
y = [x[i]**5 - 2*x[i]**2 + 0.5 for i in range(len(x))]
min_x = np.argmin(y)
print('x значение, при котором y минимально: ', x[min_x])
min_y = min(y)
print('минимальное значение функции: ', y[min_x])
max_x = np.argmax(y)
print('x значение, при котором y максимально: ', x[max_x])
print('максимальное значение функции: ', y[max_x])
```

Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId

```python
import pandas as pd
import datetime 
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
def to_datetime(a):
    return datetime.datetime.fromtimestamp(a)
```

```python
ratings['timestamp'] = ratings['timestamp'].apply(to_datetime)
```

```python
ratings.head()
ratings.info()
ratings['delta_time'] = ''
```

```python
def prepare_array(ratings):

    del_array = []

    for name, group_data in ratings.groupby('userId'):
        if group_data.shape[0] > 100:
            x_max = group_data['timestamp'].max()
            x_min = group_data['timestamp'].min()
            
            delta_time = x_max - x_min
            delta_time_str = str(delta_time)
            
            mask = (ratings['userId'] == name)            
            ratings.loc[mask, 'delta_time'] = delta_time_str
            del_array.append(name)     
       
        
    ratings = ratings.query('userId in @del_array')
    
    return ratings
```

```python
pr_ratings = prepare_array(ratings)
pr_ratings.groupby(['userId']).head()
```

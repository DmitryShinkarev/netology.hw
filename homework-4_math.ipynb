```python
from scipy.misc import derivative
def f(x):
    return 60*x**2+27*np.sin(3*x+3)-32*np.cos(2*x)
derivative(f, 1.0, dx=1e-6)
```

```python
x = np.linspace(-10, 10)
fx = f(x)
f1x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]
```

```python
plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='derivative')
plt.legend()
plt.show()
```

```python
from scipy.misc import derivative
def f(x):
    return 120*x + 64*np.sin(2*x) + 81*np.cos(3*x + 3)
derivative(f, 1.0, dx=1e-6)
```

```python
x = np.linspace(-10, 10)
fx = f(x)
f1x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]
```

```python
plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='derivative')
plt.legend()
plt.show()
```
